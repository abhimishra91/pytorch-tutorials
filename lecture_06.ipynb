**Logistic Regression**

* This is an extension to Linear model. We only have to add a `sigmoid function` on the output of our linear model that we have.

* The sigmoid will define the threshold to move the output into 0 or 1

* The loss will change from `MSE` to `Binary Cross Entropy`

```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import torch.nn.functional as F
```

```python
x_data = torch.Tensor([[1.0], [2.0], [3.0], [4.0]])
y_data = torch.Tensor([[0.0], [0.0], [1.0], [1.0]])
```

```python
# Creating the model class and defining the elements of the network

class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear = torch.nn.Linear(1,1)
        
    def forward(self, x):
        y_pred = F.sigmoid(self.linear(x))
        return y_pred

# Defining the model object for the class
model = Model()


# Defining the loss and the optimizer function
criterion = torch.nn.BCELoss(size_average=True)
optimus = torch.optim.SGD(model.parameters(), lr = 0.2)


# Training loop
for epoch in range(100):
    # forward pass
    y_pred = model(x_data)
    
    # compute loss and print
    loss = criterion(y_pred, y_data)
    print(f'Epoch: {epoch}, Loss:  {loss.item()}')
    
    # compute backward loop and update the values using optimizer
    optimus.zero_grad()
    loss.backward()
    optimus.step()
    
    
# After training
one_val = torch.Tensor([1.0])
seven_val = torch. Tensor([7.0])
print('Predict for 1 hour', 1.0, model(one_val).item()> 0.5)
print('Predict for 7 hour', 7.0, model(seven_val).item() > 0.5)
```
